# Visualizing The Gender Gap Across College Degrees

In the last project we looked at how college majors are related to income and unemployment and we noticed that median income for women was significantly less than for men, and that women were under represented in high income majors. This led to an interest in seeing if any steps have been taken to close the pay gap, which is related to the participation gap in high paying majors. This project is therefore a natural follow on from the questions that arose from the last.

The Department of Education Statistics releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970. This particular data set runs until 2012 and is broken up into 17 categories of degrees, with each column as a separate category.

Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. We will be exploring the data set to visualize the gender gap across college degrees.

### STEM Degrees

Since we often hear about the gender gap in Science, Technology, Engineering and Mathematics(STEM) degrees and we visualized it in the last project we will begin our analysis there:

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

As can be seen above there is a significant gap between the genders in Engineering and Computer Science, with men as an overwhelming majority and very little closing of the gap since 1970.

Psychology had a brief period of being a male dominated subject in the early 1970s but has since reversed this and now shows a massive gap with women dominating the field.

Biology, the Physical Sciences and Mathematics have smaller but still significant gaps with Biology having crossed the gender gap barrier in the late 1980s and is now a female dominated field.

### All Degrees

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences',
             'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 
                 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 
              'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(16, 16))

for sp in range(0,18):
    if sp%3 == 0:
        ax = fig.add_subplot(6,3,sp+1)
        ax.plot(women_degrees['Year'], (women_degrees[stem_cats[int(sp/3)]]), c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], (100-women_degrees[stem_cats[int(sp/3)]]), c=cb_orange, label='Men', linewidth=3)
    
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(stem_cats[int(sp/3)])
        ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom ='off')
        ax.set_yticks([0,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        
    if sp == 15:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
        ax.tick_params(labelbottom = 'on')
    elif sp == 0:
        ax.text(2005, 82, 'Men')
        ax.text(2001, 14, 'Women')

        
for sp in range(0,15):
    if sp%3 == 1:
        ax = fig.add_subplot(6,3,sp+1)
        ax.plot(women_degrees['Year'], (women_degrees[lib_arts_cats[int((sp-1)/3)]]), c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], (100-women_degrees[lib_arts_cats[int((sp-1)/3)]]), c=cb_orange, label='Men', linewidth=3)
    
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(lib_arts_cats[int((sp-1)/3)])
        ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom ='off')
        ax.set_yticks([0,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        
    if sp == 1:
        ax.text(2002, 77, 'Women')
        ax.text(2005, 20, 'Men')
    if sp == 14:
        ax.tick_params(labelbottom = 'on')
      
        
for sp in range(0,18):
    if sp%3 == 2:
        ax = fig.add_subplot(6,3,sp+1)
        ax.plot(women_degrees['Year'], (women_degrees[other_cats[int((sp-2)/3)]]), c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], (100-women_degrees[other_cats[int((sp-2)/3)]]), c=cb_orange, label='Men', linewidth=3)
    
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(other_cats[int((sp-2)/3)])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom ='off')
        ax.set_yticks([0,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 2:
        ax.text(2005, 7, 'Men')
        ax.text(2002, 90, 'Women')
    elif sp == 17:
        ax.text(2005, 63, 'Men')
        ax.text(2002, 33, 'Women')
        ax.tick_params(labelbottom = 'on')
        
fig.savefig('gender_degrees.png')        
plt.show()
```

Above we can see how the gender gap has shifted over the years within the different majors which have been grouped column wise into STEM fields (column 1), Liberal Arts Majors (Column 2) and Other Majors (Column 3)

With the exception of Computer Science, every category that was male dominated, that is, men constituted more than 50% of the field, has seen varying degrees of closing of the gender gap from very small in Engineering to a complete reversal as in Biology and Psychology. However the same cannot be said of historically female dominated fields which have remained largely unchanged over the years.

One may be tempted to think that since the closing gender gap has seen gains made almost exclusively by women, that it is unfair to men but that would ignore the fact that the workforce was originally disproportionately male until the advent of World War II. Moreover having seen the income numbers within the different majors it stands to reason that it is desirable for women to break into traditionally "male" fields, whereas no such incentive exists for men going into traditionally "female" fields.

Computer Science and Engineering which are the highest paying fields still show a massive gender gap in favor of men and in those particular fields there seems to be little to no progress being made to intergrate women and more needs to be done to correct this.
